```python
for i in df['duration'][:50]:
    print('min',int(i.split(':')[0]))
    print('sec',int(i.split(':')[1]))
```

```python
duration_list = []
for i in df['duration']:
    m = int(i.split(':')[0])
    s = int(i.split(':')[1])
    sec = (m*60 + s)
    duration_list.append(sec)
df['duration'] = duration_list
df.head()
```

```python
data = []

uni_singer = list(set(all_singer))
for i in tqdm(uni_singer):
    c = 0
    for singers in df['singer']:
        if i in singers:
            c += 1
    data.append([i,c])       
```
